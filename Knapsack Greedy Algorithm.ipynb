```python
def insersion_sort(L, sort_on):
    n = len(L)
    for i in range(1,n):
        key = L[i]
        j = i-1
        while j >= 0 and key[sort_on] < L[j][sort_on]:
            L[j + 1] = L[j]
            j= j-1
        L[j + 1] = key
    return L
```

```python
def knapsack(profit, weight, max_capacity, mode, knapsack_type): # mode : 1 for max profit, 2 for min weight,
    #3 for max ratio and 4 for all three of them
    n = len(weight)
    index  =list(range(1,n+1)) #list of index of the item
    ratio = [profit[p]/weight[p] for p in range(0,n)]
    
    item_tuple=zip(index, profit, weight, ratio)
    item_list = list(item_tuple)

    if mode ==1:# max profit
        sorted_items = insersion_sort(item_list,1)[::-1]
    elif mode ==2:
        sorted_items =insersion_sort(item_list,2)
    elif mode == 3:
        sorted_items = insersion_sort(item_list,3)[::-1]
    elif mode ==4:
        for m in [1,2,3]:
            knapsack(profit, weight, max_capacity, m ,knapsack_type)
        return
        
    
    index_sorted, profit_sorted, weight_sorted , ratio_sorted = map(list,(zip(*sorted_items)))
    
    total_weight = 0
    total_profit = 0
    used_items =[]
    fractions = []
    
    for i in range(0,n):
        if max_capacity ==0:
            break
        if weight_sorted[i] <= max_capacity:
            fractions.append(1)
            total_profit += profit_sorted[i]
            max_capacity -= weight_sorted[i]
            used_items.append(sorted_items[i])
        else:
            if knapsack_type ==0: #whole
                continue
                
            elif knapsack_type ==1: #fractional
                fractions.append( max_capacity/weight_sorted[i])
                total_profit+= profit_sorted[i]*fractions[i]
                used_items.append(sorted_items[i])
                break
    print_knapsack(used_items , fractions, total_profit, mode,  knapsack_type)
    #return used_items , fractions, total_profit    
        
        
    
```
